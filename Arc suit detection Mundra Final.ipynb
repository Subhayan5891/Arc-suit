```python
from ultralytics import YOLO
import cvzone
import cv2
import math
from pygame import mixer

import csv
import time
from datetime import datetime



def write_in_csv(rows):
    with open (r'YOUR_CSV_FILE.csv','r+',newline='') as file:
        file_write=csv.writer(file)
        for val in rows:
            file.write.writerow(val)



mixer.init()
mixer.music.load("download.mp3")


# Running real time from webcam
#cap = cv2.VideoCapture('buildingworker.mp4')
cap=cv2.VideoCapture(0)
#cap=cv2.VideoCapture("rtsp://tapoadmin:sharada2022@192.168.0.186/stream2/")# PKR Home
#model = YOLO('yolov8s.pt')
#model = YOLO('best.pt')
model = YOLO('base_yolov8n.pt')
#model = YOLO('best_arcsuite_mp_bg.pt')




# Reading the classes
#classnames = ['Arcsuite','No_Arcsuite']
names=['No-arcsuite', 'Arcsuite']
#classnames=['Helmet','Safety_Jacket','No_Helmet','No_Safety_Jacket']
#names = ['Hardhat', 'Mask', 'NO-Hardhat', 'NO-Mask', 'NO-Safety Vest', 'Person', 'Safety Cone',
#    'Safety Vest', 'machinery', 'vehicle']

#names = ['Helmet', 'Mask', 'NO-Helmet', 'NO-Mask', 'NO-Safety Jacket', 'Person', 'Safety Cone','Safety Jacket', 'machinery', 'vehicle']



while True:
    ret,frame = cap.read()
#    frame = cv2.resize(frame,(1080,860))
    frame = cv2.resize(frame,(640,640))
    result = model(frame,stream=True)
    

    # Getting bbox,confidence and class names informations to work with
    for info in result:
        boxes = info.boxes
        for box in boxes:
            confidence = box.conf[0]
            confidence = math.ceil(confidence * 100)
            Class = int(box.cls[0])
            if confidence > 65:
                x1,y1,x2,y2 = box.xyxy[0]
                x1, y1, x2, y2 = int(x1),int(y1),int(x2),int(y2)
                cv2.rectangle(frame,(x1,y1),(x2,y2),(0,0,255),5)
                cvzone.putTextRect(frame, f'{names[Class]} {confidence}%', [x1 + 8, y1 + 100],
                                   scale=1.5,thickness=2)
                cvzone.putTextRect(frame, f'{names[Class]}', [x1 + 8, y1 + 100],
                                   scale=1.5,thickness=2)
                
#                if (names[Class]=='NO-Hardhat' and names[Class]=='NO-Safety Vest'):
                if (names[Class]=='No-arcsuite'):
                    if not mixer.music.get_busy():
                        mixer.music.play()
                        print("Detection Active")
                    
                    
                # Terminate run when "Q" pressed
    if cv2.waitKey(1) == ord("q"):
        break





    cv2.imshow('APL Mundra Arc suit Detection',frame)
    cv2.waitKey(1)
# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
